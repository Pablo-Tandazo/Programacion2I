```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
df.head()
```

```python
setosa = df[df['species'] == 'setosa']
df['petal_ratio'] = df['petal_length'] / df['petal_width']
df.groupby('species').mean()

```

```python
sepal_lengths = df['sepal_length'].values
print("Media:", np.mean(sepal_lengths))
print("Mediana:", np.median(sepal_lengths))
print("Desviaci칩n est치ndar:", np.std(sepal_lengths))
```

```python
plt.hist(df['petal_length'], bins=20, color='lightgreen')
plt.title("Histograma de Longitud del P칠talo")
plt.xlabel("Longitud")
plt.ylabel("Frecuencia")
plt.show()


df['species'].value_counts().plot(kind='bar', color='coral')
plt.title("Cantidad de registros por especie")
plt.xlabel("Especie")
plt.ylabel("Cantidad")
plt.show()


colors = {'setosa':'red', 'versicolor':'blue', 'virginica':'green'}
for specie in df['species'].unique():
    temp = df[df['species'] == specie]
    plt.scatter(temp['sepal_length'], temp['petal_length'], label=specie, color=colors[specie])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Relaci칩n entre Sepal y Petal Length")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(df['sepal_length'], color='skyblue')
axs[1].hist(df['petal_length'], color='orange')
axs[0].set_title("Longitud del s칠palo")
axs[1].set_title("Longitud del p칠talo")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'iris' de seaborn
df = sns.load_dataset('iris')
```

```python
media = np.mean(df['petal_width'])
mediana = np.median(df['petal_width'])
desviacion = np.std(df['petal_width'])

print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviaci칩n est치ndar: {desviacion:.2f}")
```

```python
# Agrupar por especie y calcular media
promedios = df.groupby('species')['sepal_width'].mean()

# Graficar
promedios.plot(kind='bar', color='skyblue')
plt.title("Promedio de Sepal Width por Especie")
plt.ylabel("Sepal Width")
plt.xlabel("Especie")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filtrar
filtro = df[df['sepal_length'] > 6.0]

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(filtro['petal_length'].values, marker='o', linestyle='-', color='green')
plt.title("Petal Length para Sepal Length > 6.0")
plt.ylabel("Petal Length")
plt.xlabel("칈ndice")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_width', y='petal_width', hue='species', style='species')
plt.title("Dispersi칩n Sepal Width vs Petal Width")
plt.grid(True)
plt.show()
```

```python
# Crear columna petal_ratio = petal_length / petal_width
df['petal_ratio'] = df['petal_length'] / df['petal_width']

# Guardar en CSV
df.to_csv("iris_modificado.csv", index=False)
print("Archivo 'iris_modificado.csv' guardado correctamente.")
```
